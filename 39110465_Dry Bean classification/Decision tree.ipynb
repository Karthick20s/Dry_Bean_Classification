```python
#importing all the modules as required
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
#extracting the dataset
data_set = pd.read_excel("C:/Users/Admin/Desktop/New folder/DryBeanDataset/Dry_Bean_Dataset.xlsx")
data_set.head()
```

```python
y = data_set['Class'] # For y, We assigned the target variable[what we are goiing to predict].
data_set.drop('Class', axis=1, inplace=True) # For feature inputs, except label[Class]
```

```python
# Split train data into train and validation data.
x_train, x_valid, y_train, y_valid = train_test_split(data_set, y, test_size=0.2, shuffle=True) 
```

```python
# defining the mode
model = DecisionTreeClassifier()
# fitting the model with the training dataset.
model.fit(x_train, y_train)
predictions = model.predict(x_valid) # making predictions with test data.
```

```python
from sklearn.metrics import classification_report # importing the required module for confusion_matrix
from sklearn.metrics import confusion_matrix      # and classification_report.
print(classification_report(y_valid, predictions))
print(confusion_matrix(y_valid, predictions)) # for checking the accuracy, recall,precision etc,
```

```python
import matplotlib.pyplot as plt
model2 = DecisionTreeClassifier(max_depth = 2)  # TO REDUCE COMPLEXITY TAKEN MAX_DEPTH AS 2
model2.fit(x_train, y_train)
plt.figure(figsize = (20,12)) # Plotting a tree
tree.plot_tree(model2,fontsize = 15)
plt.show()
```
